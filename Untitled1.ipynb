```python
import pandas as pd
```

```python
json = pd.read_json('meta_Amazon_Fashion.jsonl',lines=True)
json.head()
```

```python
json = json.drop(['store','bought_together','details'],axis=1)
json.head()
```

```python
json = json.dropna(subset=['price'])
json.head()
```

```python
json = json.drop(['categories','videos'],axis=1)
json.head()
```

```python
json.reset_index(drop=True, inplace=True)
```

```python
import re
```

```python
def extract(x):
    return re.search(r"""(https[^\']*)""",x)[0]
```

```python
json['images'] = json['images'].astype("string")
```

```python
json['images'] = json['images'].apply(extract)
json.head()
```

```python
json = json.drop(['main_category'],axis=1)
json.head()
```

```python
json = json.drop(['average_rating', 'rating_number'],axis=1)
json.head()
```

```python
json = json.rename(columns={"title": "name", "features": "categories", "price": "priceUsd", "images": "picture", "parent_asin": "id"})
json.head()
```

```python
json = json[['id','name','description','picture','priceUsd','categories']]
json.head()
```

```python
import math
```

```python
def reformat(x):
    mydict = {"currencyCode" : "USD", "units": int(x), "nanos": math.ceil((x % 1) * 100)*10000000}
    return mydict
```

```python
print(reformat(29.81))
```

```python
json['priceUsd'] = json['priceUsd'].apply(reformat)
json.head()
```

```python
json = json.drop(['PriceUsd'], axis=1)
json.head()
```

```python
json['description'] = json['description'].astype("string")
json.head()
```

```python
json['picture'] = json['picture'].astype("string")
json.head()
```

```python
result = json.to_json('amazonproducts.json',orient="records", indent=4)
```

```python
jsonsample = json.sample(n=30000,replace=False, random_state=500)
```
